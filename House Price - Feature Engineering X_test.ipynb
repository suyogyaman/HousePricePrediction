## 2) a. Handling Missing Values 

```python
data_test.isnull().sum().any # We can see many features have NAN or null values
```

#### Lets handle NAN values for CATEGORICAL VARIABLES

```python
categorical_nan = [feature for feature in data_test.columns if data_test[feature].isnull().sum()>1 and data_test[feature].dtypes=='object']
categorical_nan
```

```python
#Lets see the percentage 
for feature in categorical_nan:
    if feature in ["Alley","PoolQC","Fence"]:
        print(feature,"\t\t\t",(np.round(data_test[feature].isnull().mean(),1)*100), '\t','% missing values')
    else:
        print(feature, "\t\t",(np.round(data_test[feature].isnull().mean(),1)*100), '\t','% missing values')
```

```python
data_test[categorical_nan].head()
```

```python
##Replace missing values with a new label ( MISSING )
def replace_cat_feature(data,categorical_nan):
    dataset = data_test.copy()
    dataset[categorical_nan] = data_test[categorical_nan].fillna('Missing')
    return dataset
```

```python
#Apply above function
data_test = replace_cat_feature(data_test,categorical_nan)
```

```python
data_test[categorical_nan].head()
```

```python
data_test[categorical_nan].isnull().sum()
```

#### Lets handle NAN values for NUMERICAL VARIABLES

```python
numerical_nan = [feature for feature in data_test.columns if data_test[feature].isnull().sum()>1 and data_test[feature].dtypes!='object']
numerical_nan
```

```python
#Lets see the percentage 
for feature in numerical_nan:
    print(feature, "\t\t",(np.round(data_test[feature].isnull().mean(),1)*100), '\t','% missing values')
```

```python
data_test[numerical_nan].isnull().sum()
```

```python
##Replacing the numerical missing values with median
for feature in numerical_nan:

    median_value = data_test[feature].median()
   
    data_test[feature+'nan'] = np.where(data_test[feature].isnull(),1,0)
    
    data_test[feature].fillna(median_value,inplace=True)
```

```python
data_test.head()
```

```python
data_test[numerical_nan].isnull().sum()
```

###  2) b. Handling Temporal Variables 

```python
##Temporal variable ( eg : DateTime Variable )
numerical_feature = [feature for feature in data_test.columns if data_test[feature].dtypes!='object']
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
year_feature.remove('GarageYrBltnan')
year_feature
```

```python
for feature in year_feature:
    data_test[feature] = data_test['YrSold'] - data_test[feature]
```

## 2)  c.   Handling Numerical Variables  
        i)  Skewed Data to Log Normal Distribution

```python
data_test.head()
```

#### In our EDA, we know that few columns have skewed values. So we need to transform numerical feature to gaussian distribution

```python
data_test.columns
```

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea']
```

```python
#Before log transformation
data_test[num_features].head()
```

```python
#Changing to log normal distribution
for feature in num_features:
    data_test[feature] = np.log(data_test[feature])
```

```python
data_test[num_features].head()
```

## 2) d.   Handling Categorical Variables 

```python
#Lets see the percentage of cateogorical features
categorical_features = [feature for feature in data_test.columns if data_test[feature].dtypes == 'O']
print("Categorical Features : \n{}".format(categorical_features))
print("\n Count  : {}".format(len(categorical_features)))
```

```python
#Creating a new label "RARE VALUE" which implies it has less frequency category in that particular feature
for feature in categorical_features:
    temp = data_test.groupby(feature).count()/len(data_test)
    temp_df = temp[temp>0.01].index
    data_test[feature] = np.where(data_test[feature].isin(temp_df),data_test[feature],'Rare_Var')
```

##  2) d.   Standarize the values of Variables  

### Feature_Scaling

```python
scaling_feature=[feature for feature in data_test.columns if feature not in ['Id','SalePrice']]
len(scaling_feature)
```

```python
data_test[scaling_feature].head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data_test[scaling_feature])
```

```python
scaler.transform(data_test[scaling_feature])
```

```python
#Transform the train and test set, and add the ID and salesprice variable
data_final = pd.concat([data_test[['Id']].reset_index(drop=True),
                       pd.DataFrame(scaler.transform(data_test[scaling_feature]),
                        columns=scaling_feature)],
                       axis=1)
```
