## 2) a. Handling Missing Values 

```python
data.isnull().sum().any # We can see many features have NAN or null values
```

#### Lets handle NAN values for CATEGORICAL VARIABLES

```python
categorical_nan = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes=='object']
categorical_nan
```

```python
#Lets see the percentage 
for feature in categorical_nan:
    if feature in ["Alley","PoolQC","Fence"]:
        print(feature,"\t\t\t",(np.round(data[feature].isnull().mean(),1)*100), '\t','% missing values')
    else:
        print(feature, "\t\t",(np.round(data[feature].isnull().mean(),1)*100), '\t','% missing values')
```

```python
data[categorical_nan].head()
```

```python
##Replace missing values with a new label ( MISSING )
def replace_cat_feature(data,categorical_nan):
    dataset = data.copy()
    dataset[categorical_nan] = data[categorical_nan].fillna('Missing')
    return dataset
```

```python
#Apply above function
data = replace_cat_feature(data,categorical_nan)
```

```python
data[categorical_nan].head()
```

```python
data[categorical_nan].isnull().sum()
```

#### Lets handle NAN values for NUMERICAL VARIABLES

```python
numerical_nan = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes!='object']
numerical_nan
```

```python
#Lets see the percentage 
for feature in numerical_nan:
    print(feature, "\t\t",(np.round(data[feature].isnull().mean(),1)*100), '\t','% missing values')
```

```python
data[numerical_nan].isnull().sum()
```

```python
##Replacing the numerical missing values with median
for feature in numerical_nan:

    median_value = data[feature].median()
   
    data[feature+'nan'] = np.where(data[feature].isnull(),1,0)
    
    data[feature].fillna(median_value,inplace=True)
```

```python
data.head()
```

```python
data[numerical_nan].isnull().sum()
```

###  2) b. Handling Temporal Variables 

```python
##Temporal variable ( eg : DateTime Variable )
numerical_feature = [feature for feature in data.columns if data[feature].dtypes!='object']
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
year_feature.remove('GarageYrBltnan')
year_feature
```

```python
for feature in year_feature:
    data[feature] = data['YrSold'] - data[feature]
```

```python
data[year_feature].head()
```

## 2)  c.   Handling Numerical Variables  
        i)  Skewed Data to Log Normal Distribution

```python
data.head()
```

#### In our EDA, we know that few columns have skewed values. So we need to transform numerical feature to gaussian distribution

```python
data.columns
```

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
```

```python
#Before log transformation
data[num_features].head()
```

```python
#Changing to log normal distribution
for feature in num_features:
    data[feature] = np.log(data[feature])
```

```python
data[num_features].head()
```

## 2) d.   Handling Categorical Variables 

```python
#Lets see the percentage of cateogorical features
categorical_features = [feature for feature in data.columns if data[feature].dtypes == 'O']
print("Categorical Features : \n{}".format(categorical_features))
print("\n Count  : {}".format(len(categorical_features)))
```

```python
#Creating a new label "RARE VALUE" which implies it has less frequency category in that particular feature
for feature in categorical_features:
    temp = data.groupby(feature)['SalePrice'].count()/len(data)
    temp_df = temp[temp>0.01].index
    data[feature] = np.where(data[feature].isin(temp_df),data[feature],'Rare_Var')
```

##   d.   Handling Categorical Variables  
              i)  Rare Category Elimination
              ii) Conversion of Category to Numerical Values

#Conversion of Cateogry to Numerical Values

```python
#First lets apply it in one column to understand the process
# 1. First get the index of each category of feature.
labels_ordered_LotShape = data.groupby('LotShape')['SalePrice'].mean().sort_values().index
labels_ordered_LotShape
```

```python
# 2. Give value to these category
labels_ordered_LotShape = {k:i for i,k in enumerate(labels_ordered_LotShape,0)}
labels_ordered_LotShape
```

```python
# 3. Convert our columns to numeric value
data['LotShape']= data['LotShape'].map(labels_ordered_LotShape)
```

```python
data.head() #Check LotShape column and compare with others
```

#### Lets perform the same for all categorical features

```python
for feature in categorical_features:
    labels_ordered = data.groupby(feature)['SalePrice'].mean().sort_values().index
    labels_ordered = {k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]= data[feature].map(labels_ordered)
```

```python
data.head()
```

##  2) d.   Standarize the values of Variables  

### Feature_Scaling

```python
scaling_feature=[feature for feature in data.columns if feature not in ['Id','SalePrice'] ]
len(scaling_feature)
```

```python
data[scaling_feature].head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(data[scaling_feature])
```

```python
scaler.transform(data[scaling_feature])
```

```python
#Transform the train and test set, and add the ID and salesprice variable
data_final = pd.concat([data[['Id','SalePrice']].reset_index(drop=True),
                       pd.DataFrame(scaler.transform(data[scaling_feature]),
                        columns=scaling_feature)],
                       axis=1)
```
